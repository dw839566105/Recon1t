```python
import numpy as np
import matplotlib.pyplot as plt
import tables
```

```python
k1 = []
for j in np.arange(0.01,0.65,0.01):
    h = tables.open_file('file_+%.3f.h5' % j)
    s = []

    for i in np.arange(2,25):
        k = eval('h.root.AIC' + str(i) + '[()]')
        s.append(k)    
    s = np.array(s)
    k1.append(np.where(s==np.min(s))[0]+1)
plt.plot(k1)
```
